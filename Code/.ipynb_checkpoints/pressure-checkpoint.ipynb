# Physics-Based Derivation of Air Pressure

This document outlines a theoretical derivation of air pressure considering temperature, humidity, and (to a very limited extent) particulate matter. The derivation is based on the ideal gas law and the concept of partial pressures in a mixture of gases.

### 1. The Ideal Gas Law

For a gas (or a mixture of gases), the ideal gas law is given by:

$$ P = ρ.R.T $$
where:
- P is the pressure,
- ρ is the density of the gas,
- R is the specific gas constant,
- T is the absolute temperature (in Kelvin).

## 2. Partial Pressures in Air

Air is a mixture of dry air and water vapor. According to Dalton's law of partial pressures, the total pressure \(P\) is the sum of the partial pressures of the components:


$$ P = P_d + e $$

where:
- P_d is the partial pressure of dry air,
- e is the partial pressure of water vapor.

### 2.1 Partial Pressure of Dry Air

For dry air, using the ideal gas law:

$$ P_d = ρ_d , R_d , T $$

where:
- ρ_d is the density of dry air,
- R_d is the specific gas constant for dry air (approximately 287 J/(kg·K)).

### 2.2 Partial Pressure of Water Vapor

The water vapor pressure (e) can be related to the relative humidity (RH) and the saturation vapor pressure (e_s(T)):


$$ e = (RH/100) . e_s(T) $$


A common empirical formula for the saturation vapor pressure (with temperature (T) in °C) is:

$$ e_s(T) ~= 6.11 x 10^((7.5T)/(273.3+T)) $$


**Note:** For the ideal gas law, temperature must be in Kelvin.

## 3. Density of Water Vapor

Using the ideal gas law for water vapor, its density (ρ_v) is given by:


$$ ρ_v = e/(R_v . T) $$

where:
- ρ_v is the water vapor density (kg/m³),
- e is the water vapor pressure (Pa),
- (R_v) is the specific gas constant for water vapor (approximately 461 J/(kg·K)),
- T is the absolute temperature (K).


## 4. Inclusion of Particulate Matter (PM)

Particulate matter (PM) is usually measured in mass concentration (e.g., µg/m³). Although PM contributes additional mass to the air, its effect on pressure is generally negligible due to:
- Its small mass fraction relative to air,
- Its non-gaseous behavior (i.e., it does not exert pressure via molecular collisions like gases).

If one wishes to include the effect of PM in a very simplified manner, the total density can be written as:

$$ ρ_total = ρ_d + ρ_v + ρ_pm $$

where: 
- ρ_total is the total air density (kg/m³),
- ρ_d is the dry air density (kg/m³),
- ρ_v is the water vapor density (kg/m³),
- ρ_pm is the particulate matter density (kg/m³).

Then, the pressure could be approximated as:

$$ P_total = ρ_total x R_m x T $$

where R_m is an effective gas constant for the mixture. In practice, since ρ_pm is many orders of magnitude smaller than ρ_d and ρ_v, its impact on the total pressure is minimal.



# Hydrostatic Balance PDE for Pressure with Temperature, Humidity, and Particulate Matter

In a static atmosphere, the **hydrostatic equilibrium** is given by

$$ ∂P/∂z = -ρ×g $$

where  
- P(x,y,z) is the atmospheric pressure,  
- ρ(x,y,z) is the air density,  
- g is the gravitational acceleration, and  
- z is the vertical coordinate.

## 1. Decomposing Air Density

Air is a mixture of dry air, water vapor, and (to a very small extent) particulate matter. We write the total density as

$$ ρ_total = ρ_d + ρ_v + ρ_pm $$

### 1.1 Dry Air Density

Using the ideal gas law for dry air, we have

$$ ρ_d = P_d/(R_d x T) $$

where  
- P_d is the partial pressure of dry air,  
- R_d ≈ 287 J/(kg·K)is the specific gas constant for dry air, and  
- T is the absolute temperature (in Kelvin).

Because the total pressure is the sum of the dry air and water vapor partial pressures,

$$ P = P_d + e $$

we can write

$$ ρ_d = (P-e)/(R_d x T). $$

### 1.2 Water Vapor Density

The water vapor density is given by

$$ ρ_v = (e)/(R_v x T), $$


with R_v ≈ 461 J/(kg·K) being the specific gas constant for water vapor. The water vapor pressure e is related to the relative humidity (RH) and the saturation vapor pressure e_s(T) by

$$ e = (RH/100) . e_s(T) $$

A common empirical formula for the saturation vapor pressure (with temperature (T) in °C) is:

$$ e_s(T) ≈ 6.11 x 10^((7.5T)/(273.3+T)) $$

### 1.3 Particulate Matter (PM)

The particulate matter mass density \(\rho_{pm}\) is usually very small compared to ρ_d and ρ_v. In many cases its effect on the overall pressure can be neglected. However, for completeness, we include it:

$$ ρ_pm = (measured or estimated PM density) $$


## 2. Formulating the PDE

Substitute the densities back into the hydrostatic balance:
$$
\[
\frac{\partial P}{\partial z} = - g \, \rho_{\text{total}} = - g \left( \frac{P-e}{R_d\,T} + \frac{e}{R_v\,T} + \rho_{pm} \right).
\]
$$
Thus, the PDE for the pressure field is
$$
\[
\boxed{
\frac{\partial P(x,y,z)}{\partial z} = - g \left( \frac{P(x,y,z)-e(x,y,z)}{R_d\,T(x,y,z)} + \frac{e(x,y,z)}{R_v\,T(x,y,z)} + \rho_{pm}(x,y,z) \right)
}
\]
$$
where
$$
\[
e(x,y,z) = \left(\frac{\text{RH}(x,y,z)}{100}\right)e_s\big(T(x,y,z)\big)
\]
$$
and
$$
\[
e_s(T) \approx 6.11 \times 10^{\frac{7.5\,T}{237.3+T}}.
\]
$$
## 3. Discussion

- **Hydrostatic Approximation:**  
  This equation assumes a balance in the vertical direction. For a full spatio-temporal model, you might need to couple this with horizontal momentum, energy, and continuity equations.

- **PINN Implementation:**  
  In a physics-informed neural network, this PDE can be incorporated as a constraint in the loss function, ensuring that the network’s predicted pressure field \(P(x,y,z)\) respects hydrostatic balance given the temperature \(T\), relative humidity \(RH\) (through \(e\)), and particulate matter \(\rho_{pm}\).

- **Extensions:**  
  For dynamic problems, additional terms (e.g., time derivatives or advective terms) might be necessary. The above is a steady-state formulation focusing on vertical balance.

---

This PDE forms a physics-based constraint that can be directly added to the loss function of a PINN to enforce that the learned pressure field is physically consistent with temperature, humidity, and particulate matter distributions.
