```python
import numpy as np
import matplotlib.pyplot as plt

nx, ny = 100, 80
dx = dy = 100  
dt = 60        
nt = 200      

alpha = 0.5  
beta = 0.5   
gamma = 0.1  
k = 0.01  

# Initialize pressure and temperature fields
P = np.random.rand(ny, nx) * 2  # Random initial pressure field
T = np.zeros((ny, nx))  

for i in range(-3, 4):
    for j in range(-3, 4):
        T[ny//2 + i, nx//2 + j] = 300  

# Laplacian function
def laplacian(Z, dx, dy):
    return (
        (np.roll(Z, -1, axis=1) - 2 * Z + np.roll(Z, 1, axis=1)) / dx**2 +
        (np.roll(Z, -1, axis=0) - 2 * Z + np.roll(Z, 1, axis=0)) / dy**2
    )

# Time stepping loop
for n in range(nt):
    # Update Temperature field (heat diffusion)
    T += dt * k * laplacian(T, dx, dy)

    # Update Pressure field
    dP_dt = alpha * laplacian(T, dx, dy) + beta * laplacian(P, dx, dy) + gamma * T
    P_new = P + dt * dP_dt  

    P_new[0, :] = P_new[1, :]
    P_new[-1, :] = P_new[-2, :]
    P_new[:, 0] = P_new[:, 1]
    P_new[:, -1] = P_new[:, -2]

    P = P_new.copy()

    # Visualization every 10 steps
    if n % 10 == 0:
        plt.clf()
        plt.imshow(P, cmap='plasma', origin='lower', extent=[0, nx*dx, 0, ny*dy])
        plt.title(f"Pressure Field Evolution at t={n*dt} s")
        plt.xlabel("X (meters)")
        plt.ylabel("Y (meters)")
        plt.colorbar(label="Pressure")
        plt.pause(0.1)

plt.show()
```
